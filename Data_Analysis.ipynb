```python
# Importing Libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os
```

```python
# Connecting with dataset

dataset = pd.read_csv('Resources/world-happiness-report-2021.csv')

dataset.head()
```

```python
# Making a copy of the dataset

dataset_modified = dataset.copy()
dataset_modified
```

```python
# Checking the data types

dataset_modified.dtypes
```

```python
# Rounding to two decimal points

dataset_modified = dataset_modified.round(2)

dataset_modified.head()
```

```python
# Renaming the columns

dataset_modified = dataset_modified.rename(columns={"Country name": "Country Name", "Regional indicator": "Region", "Ladder score": "Country Score", "Logged GDP per capita": "GDP per capita", "Social support": "Social Support", "Healthy life expectancy": "Life Expectancy", "Freedom to make life choices": "Freedom", "Perceptions of corruption": "Corruption Score"})

dataset_modified.head()
```

```python
# Deleting unnecessary columns

dataset_modified = dataset_modified.drop(columns=['Standard error of ladder score', 'upperwhisker', 'lowerwhisker', 'Ladder score in Dystopia', 'Explained by: Log GDP per capita', 'Explained by: Social support', 'Explained by: Healthy life expectancy', 'Explained by: Freedom to make life choices', 'Explained by: Generosity', 'Explained by: Perceptions of corruption', 'Dystopia + residual'])

dataset_modified.head()
```

```python
# Scatter plot of Country Score vs GDP per capita

fig = plt.gcf()

fig.set_size_inches(14, 8)

# Setting the seaborn theme for the scatter plot
sns.set_style("whitegrid")

ax = sns.scatterplot(data=dataset_modified, x='Country Score', y='GDP per capita', hue='Region', size='GDP per capita').set(title='Country Score vs GDP per capita')
```

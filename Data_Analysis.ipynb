```python
# Importing Libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os
```

```python
# Connecting with dataset

dataset = pd.read_csv('Resources/world-happiness-report-2021.csv')

dataset.head()
```

```python
# Making a copy of the dataset

dataset_modified = dataset.copy()
dataset_modified
```

```python
# Checking the data types

dataset_modified.dtypes
```

```python

# Rounding to two decimal points

dataset_modified = dataset_modified.round(2)

dataset_modified.head()
```

```python

# Renaming the columns

dataset_modified = dataset_modified.rename(columns={"Regional indicator": "Region", "Ladder score": "Score", "Logged GDP per capita": "GDP per capita", "Social support": "Social Support", "Healthy life expectancy": "Life Expectancy", "Freedom to make life choices": "Freedom", "Perceptions of corruption": "Corruption Score"})

dataset_modified.head()
```

```python
# Scatter plot of Score vs GDP per capita

fig = plt.gcf()

fig.set_size_inches(14, 8)

sns.set_style("darkgrid")

sns.scatterplot(data=dataset_modified, x='Score', y='GDP per capita', hue='Region', size='GDP per capita').set(title='Score vs GDP per capita')
```

```python
# Scatter plot of Score vs Social Support

fig = plt.gcf()

fig.set_size_inches(14, 8)

sns.set_style("darkgrid")

sns.scatterplot(data=dataset_modified, x='Score', y='Social Support', hue='Region', size='Social Support').set(title='Score vs Social Support')
```
